## Reload the data


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

meteorites = pd.read_csv("data/meteorites.csv")

meteorites = meteorites.drop('Unnamed: 0', axis=1)

meteorites
```

## Preliminary Statistical Analysis

```python
plt.hist(meteorites.mass_in_grams, bins='auto', density=True, facecolor='blue',label='Discrete PDF', alpha=0.25, edgecolor="k")

plt.xlabel("mass_in_grams")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')
```
