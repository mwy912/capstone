## Reload the data


```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab

meteorites = pd.read_csv("data/meteorites.csv")

meteorites = meteorites.drop('Unnamed: 0', axis=1) # remove old ID column

meteorites
```

## Column Prep: Categories

```python
# Fell or Found:

meteorites['fell_or_found'] = pd.Categorical(meteorites['fell_or_found'])

# Size:

size = []
for row in meteorites['mass_in_grams']:
        if   row <   100: size.append('tiny')
        elif row <  1000: size.append('small')
        elif row <  5000: size.append('medium')
        elif row < 10000: size.append('large')
        else:             size.append('huge')
meteorites['size'] = size
meteorites['size']= pd.Categorical(meteorites['size'])

# Recclass

meteorites['recclass']= pd.Categorical(meteorites['recclass'])
```

## Visualization and charts

```python
fall_pie = meteorites['fell_or_found'].value_counts()

print(fall_pie)
plot = fall_pie.plot.pie(figsize=(5, 5))
```

```python
size_pie = meteorites['size'].value_counts()
print (size_pie)
plot = size_pie.plot.pie(figsize=(5, 5))
```

```python
recclass_pie = meteorites['recclass'].value_counts()

plot = recclass_pie.plot.pie(figsize=(20,20))

```

```python
recclass_pie = meteorites['recclass'].value_counts()

plot = recclass_pie.plot.barh(figsize=(20,100))
```
