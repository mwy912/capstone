```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import scipy.stats as stats
```

```python
ff_train_wt = pd.read_csv("data/results/traintest/meteorite_falls_train_weighted.csv")
ff_train_wt.drop(ff_train_wt.columns[ff_train_wt.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
ff_train_wt.drop(columns = ["X"],inplace=True)
ff_train_wt.head()
```

```python
df = ff_train_wt

plt.figure(figsize=(16, 8)) 
plt.subplot(2,2,1) 
plt.hist(df.mass, bins='auto', density=True, facecolor='blue', label='Discrete PDF', edgecolor="k")
plt.xlabel("Mass")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.subplot(2,2,2) 
plt.hist(df.year, bins='auto', density=True, facecolor='blue', label='Discrete PDF', edgecolor="k")
plt.xlabel("year")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.subplot(2,2,3) 
plt.hist(df.latitude, bins='auto', density=True, facecolor='blue', label='Discrete PDF', edgecolor="k")
plt.xlabel("latitude")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.subplot(2,2,4) 
plt.hist(df.longitude, bins='auto', density=True, facecolor='blue', label='Discrete PDF', edgecolor="k")
plt.xlabel("longitude")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.show()
```

```python
plt.figure(figsize=(16, 8)) 
plt.subplot(2,2,1) 
plt.hist(df.mass, bins='auto', density=True, facecolor='blue',label='Discrete PDF', alpha=0.25, edgecolor="k")

mu = df['mass'].mean()
sigma = df['mass'].std()
x = np.arange(0,1000000,100000)
y = stats.norm.pdf(x, mu, sigma)
# or: y = stats.norm.pdf(x, mu, sigma)
plt.plot(x,y,'r-')
plt.xlabel("mass")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.subplot(2,2,2) 
plt.hist(df.year, bins='auto', density=True, facecolor='blue',label='Discrete PDF', alpha=0.25, edgecolor="k")
mu = df['year'].mean()
sigma = df['year'].std()
x = np.arange(0,2021,50)
y = stats.norm.pdf(x, mu, sigma)
# or: y = stats.norm.pdf(x, mu, sigma)
plt.plot(x,y,'r-')
plt.xlabel("Year")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.subplot(2,2,3) 
plt.hist(df.latitude, bins='auto', density=True, facecolor='blue',label='Discrete PDF', alpha=0.25, edgecolor="k")
mu = df['latitude'].mean()
sigma = df['latitude'].std()
x = np.arange(-90,90,1)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x,y,'r-')
plt.xlabel("latitude")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.subplot(2,2,4) 
plt.hist(df.longitude, bins='auto', density=True, facecolor='blue',label='Discrete PDF', alpha=0.25, edgecolor="k")
mu = df['longitude'].mean()
sigma = df['longitude'].std()
x = np.arange(-180,180,20)
y = stats.norm.pdf(x, mu, sigma)
plt.plot(x,y,'r-')
plt.xlabel("longitude")
plt.ylabel("Probability Density")
plt.legend(loc='upper right')

plt.show()
```
