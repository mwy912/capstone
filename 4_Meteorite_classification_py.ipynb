## Meteorite Classification

```python
import pandas as pd
import numpy as np

meteoritegroups = pd.read_csv("data/class/group_table.csv")
meteoritegroups
```

```python
#Bring in meteorites table

meteorites = pd.read_csv("data/results/meteorites.csv")
meteorites = meteorites.drop('Unnamed: 0', axis=1) # remove old ID column

meteorites
```

```python
#make a new column to clean up recclass so I can join it to the meteorclass table.

#first remove numbers and add to new column

meteorites["group"] = meteorites["recclass"].str.replace('\d+', '')

meteorites
```

```python
# checking the unique values left.

uclass_array = meteorites['group'].unique()

uclass_l = uclass_array.tolist()

uclass_l = sorted(uclass_l)

uclass_l
```

```python
#list of meteorites classes.

all_groups_array = meteoritegroups['group'].unique()
all_groups_l = all_groups_array.tolist()
all_groups_l = sorted(all_groups_l)
all_groups_l
```
