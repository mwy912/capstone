```r
## Load relevant R packages.

inLibraries = list('repr','spdep','raster','rgdal','rgeos')
for (rpack in inLibraries) {
  if (is.element(rpack,installed.packages()[,1])){
      #Load the library into R
      suppressMessages(library(rpack,character.only = TRUE))
    }
    else {
        print(paste("Warning:  ",rpack," is not an installed package"))
    }
}

##  Resize plot area.

options(repr.plot.width=6, repr.plot.height=6)
```

## Population

```r
pc1990name <- 'data/pop/popcount/gpw_pop_counts_hd_1990.asc' 
pc1990 <- raster(pc1990name,values=TRUE)

pc1995name <- 'data/pop/popcount/gpw_pop_counts_hd_1995.asc' 
pc1995 <- raster(pc1995name,values=TRUE)

plot(pc1990,main="Population, 1990")
plot(pc1995,main="Population, 1995")
```

## Population Density

```r
pd1990name <- 'data/pop/popdens/gpw_pop_density_hd_1990.asc'
pd1990 <- raster(pd1990name,values=TRUE)

pd1995name <- 'data/pop/popdens/gpw_pop_density_hd_1995.asc' 
pd1995 <- raster(pd1995name,values=TRUE)

plot(pd1990,main="Population Density, 1990")
plot(pd1995,main="Population Density, 1995")
```
