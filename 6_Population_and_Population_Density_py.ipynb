## Population

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#Population

fname = 'data/pop/popcount/gpw_pop_counts_hd_1990.asc'
pc1990 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pc1990, interpolation='nearest')
plt.title('Population in 1990')
plt.show()

fname = 'data/pop/popcount/gpw_pop_counts_hd_1995.asc'
pc1995 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pc1995, interpolation='nearest')
plt.title('Population in 1995')
plt.show()
```

## Population Density

```python
#Population Density

fname = 'data/pop/popdens/gpw_pop_density_hd_1990.asc'
pd1990 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pd1990, interpolation='nearest')
plt.title('Population Density in 1990')
plt.show()

fname = 'data/pop/popdens/gpw_pop_density_hd_1995.asc'
pd1995 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pd1995, interpolation='nearest')
plt.title('Population Density in 1995')
plt.show()
```
