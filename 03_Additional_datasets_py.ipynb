## Landcover Data

Data: ISLSCP II Historical Land Cover and Land Use, 1700-1990

Source:  https://daac.ornl.gov/ISLSCP_II/guides/historic_landcover_xdeg.html

```python
import numpy as np

fname = "data/lc/historic_landcover_hd_1700.asc"
lc1700 = np.loadtxt(fname, skiprows = 6)

fname = "data/lc/historic_landcover_hd_1750.asc"
lc1750 = np.loadtxt(fname, skiprows = 6)

fname = "data/lc/historic_landcover_hd_1800.asc"
lc1800 = np.loadtxt(fname, skiprows = 6)

fname = "data/lc/historic_landcover_hd_1850.asc"
lc1850 = np.loadtxt(fname, skiprows = 6)

fname = "data/lc/historic_landcover_hd_1900.asc"
lc1900 = np.loadtxt(fname, skiprows = 6)

fname = "data/lc/historic_landcover_hd_1950.asc"
lc1950 = np.loadtxt(fname, skiprows = 6)

fname = "data/lc/historic_landcover_hd_1970.asc"
lc1970 = np.loadtxt(fname, skiprows = 6)

fname = "data/lc/historic_landcover_hd_1990.asc"
lc1990 = np.loadtxt(fname, skiprows = 6)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.imshow(lc1700, interpolation='nearest')
plt.title('Land Cover in 1700')
plt.show()

plt.imshow(lc1750, interpolation='nearest')
plt.title('Land Cover in 1750')
plt.show()

plt.imshow(lc1800, interpolation='nearest')
plt.title('Land Cover in 1800')
plt.show()

plt.imshow(lc1850, interpolation='nearest')
plt.title('Land Cover in 1850')
plt.show()

plt.imshow(lc1900, interpolation='nearest')
plt.title('Land Cover in 1900')
plt.show()

plt.imshow(lc1950, interpolation='nearest')
plt.title('Land Cover in 1950')
plt.show()

plt.imshow(lc1970, interpolation='nearest')
plt.title('Land Cover in 1970')
plt.show()

plt.imshow(lc1990, interpolation='nearest')
plt.title('Land Cover in 1990')
plt.show()
```

## Population and Population Density

Data: ISLSCP II Global Population of the World 1990-1995

Source: https://daac.ornl.gov/cgi-bin/dsviewer.pl?ds_id=975

```python
#Population

fname = 'data/pop/popcount/gpw_pop_counts_hd_1990.asc'
pc1990 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pc1990, interpolation='nearest')
plt.title('Population in 1990')
plt.show()

fname = 'data/pop/popcount/gpw_pop_counts_hd_1995.asc'
pc1995 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pc1995, interpolation='nearest')
plt.title('Population in 1995')
plt.show()


#Population Density

fname = 'data/pop/popdens/gpw_pop_density_hd_1990.asc'
pd1990 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pd1990, interpolation='nearest')
plt.title('Population Density in 1990')
plt.show()

fname = 'data/pop/popdens/gpw_pop_density_hd_1995.asc'
pd1995 = np.loadtxt(fname, skiprows = 6)

plt.imshow(pd1995, interpolation='nearest')
plt.title('Population Density in 1995')
plt.show()


```
